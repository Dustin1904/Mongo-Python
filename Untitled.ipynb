```python
import pymongo
import pandas as pd
from pymongo import MongoClient

# Conectar a MongoDB
client = MongoClient('127.0.0.1', 27017)
db = client['DeberAnalisis']
mi_coleccion = db['Base']

# Cargar el archivo CSV en un DataFrame
archivo_csv = 'Fortnite Statistics.csv'
df = pd.read_csv(archivo_csv)

# Convertir el DataFrame a una lista de diccionarios (cada fila es un diccionario)
datos_para_insertar = df.to_dict(orient='records')

# Insertar los documentos en la colección
mi_coleccion.insert_many(datos_para_insertar)
```
